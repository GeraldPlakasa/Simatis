```python
makanan = {"Rendang":[20,62], "Sate":[15,57], "Nasi Goreng":[10,52], "Steak":[22,64], "Bakwan":[5,47], "Ayam Goreng":[7,49]}

def ageUpList(karakter_list):
    for char in karakter_list:
        char.age += 1

def ageUp(karakter):
    if karakter.age < 10:
        karakter.coin += 500
    karakter.age += 1
    return karakter

def gaji(karakter_list):
    for i in range(len(karakter_list)-1):
        if karakter_list[i].role == "King" or karakter_list[i].role == "Queen":
            karakter_list[i].coin += 2500
        elif karakter_list[i].role == "Panglima":
            karakter_list[i].coin += 1000
        elif karakter_list[i].role == "Prajurit":
            karakter_list[i].coin += 500
        else:
            karakter_list[i].coin += 100

def daily(karakter_list):
    for char in karakter_list:
        char.setSexRole()
        temp_stamina = char.power.stamina
        karakter_atk = [char2.power.atk for char2 in karakter_list]
        rerata_atk = int(sum(karakter_atk)/len(karakter_atk))
        # Di ulang terus sampai stamina habis
        while(temp_stamina > 0):

            if char.batas_hungry < 0:
                char.batas_hungry = 0
            elif char.batas_hungry < 30:
                char.batas_hungry = char.makan()
            elif char.batas_hungry > 100:
                char.batas_hungry = 100
            else:
                if char.power.stamina < 25:
                    stamina = char.power.stamina
                    hungry = 1
                else:
                    stamina = random.randint(25, char.power.stamina)
                    hungry = random.randint(10, 30)

                temp_stamina -= stamina
                char.batas_hungry -= hungry

                kegiatan = random.randint(0, 100)

        #         if char.power.atk > rerata_atk:
        #           kegiatan = random.randint(0, 100)
        #         else:
        #           kegiatan = random.randint(11, 100)
        
                # Jika umur di bawah 10 hanya latihan jika tidak random
                if char.age < 15:
                    if kegiatan <= 50:
                        char.latih()
                    else:
                        pass
                else:
                    if kegiatan <= 1:
                        char.foyaFoya()
                    elif kegiatan <= 6:
                        char.boss()
                    elif kegiatan <= 36:
                        char.latih()
                    elif kegiatan <= 66:
                        char.hunt()
                    else:
                        pass

    return karakter_list

def duel(karakterA, karakterB):
    healthA = karakterA.power.health
    healthB = karakterB.power.health

    if(karakterA.power.spd >= karakterB.power.spd):
        while(healthA > 0 and healthB > 0):
            dmg = karakterA.serang(karakterB)
            healthB -= dmg
            dmg = karakterB.serang(karakterA)
            healthA -= dmg
    else:
        while(healthA > 0 and healthB > 0):
            dmg = karakterB.serang(karakterA)
            healthA -= dmg
            dmg = karakterA.serang(karakterB)
            healthB -= dmg
      
    if(healthA <= 0):
        # print(karakterB.name+" Winner")
        # exp_duel = random.randint(30, 50)
        # ExpUp(karakterB, exp_duel)
        # exp_duel = random.randint(10, 30)
        # ExpUp(karakterA, exp_duel)
        return karakterB.name
    else:
        # print(karakterA.name+" Winner")
        # exp_duel = random.randint(30, 50)
        # ExpUp(karakterA, exp_duel)
        # exp_duel = random.randint(10, 30)
        # ExpUp(karakterB, exp_duel)
        return karakterA.name

def ExpUp(karakter, exp_new):
    karakter.power.exp += exp_new
    if karakter.power.exp >= karakter.power.batas_exp:
        levelUp(karakter)

def levelUp(karakter):

  # get new Stats
    if karakter.age < 10:
        health_new = random.randint(10, 50)
        spd_new = random.randint(1, 5)
        defend_new = random.randint(1, 8)
        atk_new = random.randint(1, 8)
        stamina_new = random.randint(0, 2)
    else:
        health_new = random.randint(40, 90)
        spd_new = random.randint(2, 10)
        defend_new = random.randint(8, 20)
        atk_new = random.randint(8, 20)
        stamina_new = random.randint(2, 10)

    # plus new Stats
    karakter.power.health += health_new
    karakter.power.spd += spd_new
    karakter.power.defend += defend_new
    karakter.power.atk += atk_new
    karakter.power.stamina += stamina_new

    karakter.level += 1
    karakter.power.batas_exp += 100
    karakter.power.exp = 0

def cariPasangan(karakter_list):
    
    karakter_list_berumur = []
    karakter_list_berumur = [char for char in karakter_list if char.age >= 18 and char.age <= 50 and char.siap_pasangan == False]
    if len(karakter_list_berumur) > 0:
        karakter_siap = random.choices(karakter_list_berumur, k=math.ceil(len(karakter_list_berumur)*0.01))
        karakter_list_berumur_p = [char for char in karakter_siap if char.sex == "Perempuan"]
        karakter_list_berumur_l = [char for char in karakter_siap if char.sex == "Laki-laki"]
        banyak_pasangan = min([len(karakter_list_berumur_p), len(karakter_list_berumur_l)])
        if banyak_pasangan > 0:
            # keluarga_list
            for i in range(banyak_pasangan):

                perempuan = random.choices(karakter_list_berumur_p)[0]
                karakter_list_berumur_p.remove(perempuan)
                laki = random.choices(karakter_list_berumur_l)[0]
                karakter_list_berumur_l.remove(laki)
                laki.pasangan = perempuan.name
                laki.siap_pasangan = True
                perempuan.pasangan = laki.name
                perempuan.siap_pasangan = True
                keluarga_list.append(Keluarga(laki, perempuan))

def updateKeluarga(karakter_list, keluarga_list, list_girl, list_boy):
    for keluarga in keluarga_list:
        int_anak = random.randint(0, 100)
        if len(keluarga.anak) >= 3:
            batas_persen = 5
        else:
            batas_persen = 20
        if int_anak < batas_persen:
            kelamin = random.randint(0, 10)
            if kelamin < 5:
                sex = "Laki-laki"
                nama = random.choices(list_boy)[0]
                list_boy.remove(nama)
            else:
                sex = "Perempuan"
                nama = random.choices(list_girl)[0]
                list_girl.remove(nama)

            karakter_list.append(keluarga.createAnak(nama, 1, sex))
    return karakter_list, keluarga_list, list_girl, list_boy

def dead(karakter_list):
    int_mati = random.randint(0, 100)
    if int_mati < 1:
        jumlah_mati = random.randint(1, 5)
        for i in range(jumlah_mati):
            char_mati = random.choices(karakter_list)[0]
            karakter_list.remove(char_mati)
            print(char_mati.name + " Mati Random")
        
    karakter_list_berumur = [char for char in karakter_list if char.age >= 100]
    for char in karakter_list_berumur:
        int_mati = random.randint(0, 100)
        if int_mati < 5:
            karakter_list.remove(char)
            print(char.name + " Mati!!!")

class kekuatan:

    def __init__(self, age):
        self.atk = random.randint(1, 20)
        self.defend = random.randint(0, 20)
        self.health = random.randint(200, 1000)
        self.spd = random.randint(0, 10)
        self.exp = 0
        self.batas_exp = 100
        if age < 15:
            self.stamina = int(random.randint(25, 75) * (age - 1)/(14 - 1)) + 1
        elif age >= 15 and age <= 40 :
            self.stamina = random.randint(75, 150)
        else:
            self.stamina = int((random.randint(25, 75) / ((age - 40)/(100 - 40))) / 10)

class karakter:

    def __init__(self, name, age, sex):
        self.name = name
        if age <= 0:
            age = 1
        elif age > 100:
            age = 100
        self.age = age
        
        self.setSexRole()
        
        self.power = kekuatan(age)
        self.level = 0
        self.coin = 750
        self.batas_hungry = 100
        self.sex = sex
        self.role = "Penduduk"
        
        self.pasangan = None
        self.siap_pasangan = False
    
    def setSexRole(self):
        if self.age < 7:
            self.sex_role = "Anak-anak"
        elif self.age < 15:
            self.sex_role = "Remaja"
        elif self.age < 30:
            self.sex_role = "Dewasa Muda"
        elif self.age < 60:
            self.sex_role = "Dewasa Tua"
        else:
            self.sex_role = "Lansia"
  
    def latih(self):
        if self.age < 10:
              exp_latihan = random.randint(2, 10)
        else:
              exp_latihan = random.randint(10, 30)
        ExpUp(self, exp_latihan)
  
    def hunt(self):
        self.coin += random.randint(50, 100)
        exp_hunt = random.randint(5, 40)
        ExpUp(self, exp_hunt)

    def boss(self):
        self.coin += random.randint(500, 700)
        exp_boss = random.randint(50, 70)
        ExpUp(self, exp_boss)

    def foyaFoya(self):
        self.coin -= random.randint(500, 2000)

    def serang(self, musuh):
        dmg = self.power.atk*(100/(100+musuh.power.defend)) * random.uniform(0.75, 1.25)
        critical = random.randint(0, 100)
        miss = random.randint(0, 100)
        if critical <= 20:
              dmg *= 2
        if miss <= 20:
              dmg *= 0
        return dmg

    def makan(self):
        keys = list(makanan.keys())
        makan_pilihan = random.randint(0, len(keys)-1)
        if self.age > 10:
            self.batas_hungry += (makanan.get(keys[makan_pilihan])[1] - 30)
            self.coin -= (makanan.get(keys[makan_pilihan])[0] + 20)
        else:
            self.batas_hungry += makanan.get(keys[makan_pilihan])[1]
            self.coin -= makanan.get(keys[makan_pilihan])[0]
        return self.batas_hungry

    def info(self):
        print("Name : "+self.name)
        print("Age : "+str(self.age))
        print("Kelamin : "+self.sex)
        print("sex_Role : "+self.sex_role)
        print("Pasangan : "+str(self.pasangan))
        print("Level : "+str(self.level))
        print("Coin : "+str(self.coin))
        print("ATK : "+str(self.power.atk))
        print("DEF : "+str(self.power.defend))
        print("SPD : "+str(self.power.spd))
        print("Health : "+str(self.power.health))
        print("Stamina : "+str(self.power.stamina))
        print("EXP : "+str(self.power.exp))
        print(" ")

class Keluarga:
    
    def __init__(self, suami, istri):
        self.suami = suami
        self.istri = istri
        self.anak = []
        
    def suamiInfo(self):
        self.suami.info()
    def istriInfo(self):
        self.istri.info()
    def anakInfo(self):
        for ank in self.anak:
            ank.info()
            print("-----------")
            
    def createAnak(self, nama, umur, sex):
        anak = karakter(nama, umur, sex)
        self.anak.append(anak)
        return anak
```
