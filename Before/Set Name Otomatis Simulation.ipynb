```python
import random
import pandas as pd
```

```python
df = pd.read_csv("NationalNames.csv")
df
```

```python
df_girl = df.loc[df["Gender"] == "F"][["Name"]]
df_girl = df_girl.drop_duplicates()
df_girl = df_girl.reset_index()
df_girl = df_girl.drop(["index"], axis=1)
df_girl
```

```python
df_boy = df.loc[df["Gender"] == "M"][["Name"]]
df_boy = df_boy.drop_duplicates()
df_boy = df_boy.reset_index()
df_boy = df_boy.drop(["index"], axis=1)
df_boy
```

```python
df_indo = pd.read_csv("indonesian-names.csv")
df_indo
```

```python
df_cewe = df_indo.loc[df_indo["gender"] == "f"][['name']]
df_cewe = df_cewe.drop_duplicates()
df_cewe = df_cewe.reset_index()
df_cewe = df_cewe.drop(["index"], axis=1)
df_cewe.rename(columns={'name': 'Name'}, inplace=True)
df_cewe
```

```python
df_cowo = df_indo.loc[df_indo["gender"] == "m"][['name']]
df_cowo = df_cowo.drop_duplicates()
df_cowo = df_cowo.reset_index()
df_cowo = df_cowo.drop(["index"], axis=1)
df_cowo.rename(columns={'name': 'Name'}, inplace=True)
df_cowo
```

```python
df_p = pd.concat([df_girl, df_cewe])
df_p = df_p.drop_duplicates()
df_p = df_p.reset_index()
df_p = df_p.drop(["index"], axis=1)
df_p
```

```python
df_l = pd.concat([df_boy, df_cowo])
df_l = df_l.drop_duplicates()
df_l = df_l.reset_index()
df_l = df_l.drop(["index"], axis=1)
df_l
```

```python
df_p.to_csv("nama_girl.csv")
df_l.to_csv("nama_boy.csv")
```
